```python
import pandas as pd
import numpy as np


df = pd.read_csv("Student.csv")
print(" Student Data:")
print(df)


df["Average"] = df[["Maths", "Science", "English"]].mean(axis=1)
print("\n Average Marks:")
print(df[["Name", "Average"]])


print("\n Subject Wise Statistics:")
print(df.describe())


topper = df.loc[df["Average"].idxmax()]
print("\n Class Topper:")
print(topper["Name"], "with Average Marks:", topper["Average"])


marks_array = df[["Maths", "Science", "English"]].to_numpy()
overall_mean = np.mean(marks_array)
print("\n Overall Mean of all marks:", overall_mean)


```
